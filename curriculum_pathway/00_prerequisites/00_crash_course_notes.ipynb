## 📐 Mathematical Foundations

1. **Matrix Multiplication**: If $A$ is of size $(m, n)$ and $B$ is of size $(n, p)$, their product $C = AB$ is of size $(m, p)$:
   $$C_{ij} = \sum_{k=1}^n A_{ik} B_{kj}$$
2. **Transpose**: If $X$ is shape $(m, n)$, then $X^T$ is shape $(n, m)$, where $(X^T)_{ij} = X_{ji}$.
3. **Derivatives**: For a function $f(w) = w^2$, the derivative $f'(w) = 2w$. For a vector weight formulation $f(\mathbf{w}) = \mathbf{w}^T \mathbf{w}$, the gradient $\nabla_\mathbf{w} f(\mathbf{w}) = 2\mathbf{w}$.
