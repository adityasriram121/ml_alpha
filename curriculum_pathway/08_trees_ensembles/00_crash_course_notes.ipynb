## 📐 Mathematical Foundations
1. **Gini Impurity**:
   $$G_i = 1 - \sum_{k=1}^K p_{i, k}^2$$
2. **Entropy**:
   $$H_i = -\sum_{k=1}^K p_{i, k} \log_2(p_{i, k})$$
3. **Random Forest Feature Selection**: At each node split, only a random subset of $m \approx \sqrt{n}$ features is considered.
