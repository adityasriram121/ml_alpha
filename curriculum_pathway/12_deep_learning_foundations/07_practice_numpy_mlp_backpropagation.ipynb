## Mathematical Architecture

Our network has 1 input layer ($n_{in}$ features), 1 hidden layer ($n_{hid}$ units), and 1 output layer ($1$ unit with Sigmoid activation).

### 1. Forward Pass
For feature matrix $X$ of shape $(m, n_{in})$:
$$
Z^{[1]} = X W^{[1]} + b^{[1]} \quad \text{shape: } (m, n_{hid})
$$
$$
A^{[1]} = \text{ReLU}(Z^{[1]}) \quad \text{shape: } (m, n_{hid})
$$
$$
Z^{[2]} = A^{[1]} W^{[2]} + b^{[2]} \quad \text{shape: } (m, 1)
$$
$$
A^{[2]} = \sigma(Z^{[2]}) = \frac{1}{1 + e^{-Z^{[2]}}} \quad \text{shape: } (m, 1)
$$

### 2. Loss Function (Binary Cross-Entropy)
$$
L = -\frac{1}{m} \sum_{i=1}^m \left[ Y^{(i)} \log(A^{[2](i)}) + (1 - Y^{(i)}) \log(1 - A^{[2](i)}) \right]
$$

### 3. Backward Pass (Gradients)
Using the chain rule, we compute derivatives of $L$ with respect to each parameter. Let $dZ = \frac{\partial L}{\partial Z}$:
$$
dZ^{[2]} = A^{[2]} - Y \quad \text{shape: } (m, 1)
$$
$$
dW^{[2]} = \frac{1}{m} (A^{[1]})^T dZ^{[2]} \quad \text{shape: } (n_{hid}, 1)
$$
$$
db^{[2]} = \frac{1}{m} \sum_{i=1}^m dZ^{[2](i)} \quad \text{shape: } (1, 1)
$$
$$
dA^{[1]} = dZ^{[2]} (W^{[2]})^T \quad \text{shape: } (m, n_{hid})
$$
$$
dZ^{[1]} = dA^{[1]} \odot \text{ReLU}'(Z^{[1]}) \quad \text{shape: } (m, n_{hid})
$$
where $\odot$ is element-wise multiplication, and $\text{ReLU}'(Z^{[1]})$ is $1$ if $z > 0$ else $0$.
$$
dW^{[1]} = \frac{1}{m} X^T dZ^{[1]} \quad \text{shape: } (n_{in}, n_{hid})
$$
$$
db^{[1]} = \frac{1}{m} \sum_{i=1}^m dZ^{[1](i)} \quad \text{shape: } (1, n_{hid})
$$
