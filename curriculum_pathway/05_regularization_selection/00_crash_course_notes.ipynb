## 📐 Mathematical Foundations
1. **L2 Penalty (Ridge)**:
   $$L_{Ridge}(\theta) = \text{MSE}(\theta) + \alpha \frac{1}{2} \sum_{j=1}^n \theta_j^2$$
2. **L1 Penalty (Lasso)**:
   $$L_{Lasso}(\theta) = \text{MSE}(\theta) + \alpha \sum_{j=1}^n |\theta_j|$$
