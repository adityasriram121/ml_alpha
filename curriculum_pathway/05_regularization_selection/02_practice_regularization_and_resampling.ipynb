## Mathematical Objectives

1. **Ridge Regression Loss (L2)**:
   $$J(\theta) = \text{MSE}(\theta) + \alpha \frac{1}{2} \sum_{j=1}^n \theta_j^2$$
   *Note: The bias term $\theta_0$ is typically NOT regularized.*

2. **Lasso Regression Loss (L1)**:
   $$J(\theta) = \text{MSE}(\theta) + \alpha \sum_{j=1}^n |\theta_j|$$

3. **K-Fold Cross-Validation**:
   * Partition dataset randomly into $K$ equal-sized folds.
   * For each fold $i = 1 \dots K$:
     * Train the model on all folds *except* fold $i$.
     * Evaluate the model (compute MSE) on fold $i$.
   * Return the average MSE over all $K$ folds.
