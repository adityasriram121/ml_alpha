## 📐 Mathematical Foundations

1. **Mean Squared Error (MSE)**: Measures prediction quality:
   $$\text{MSE} = \frac{1}{m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)})^2$$
2. **Bias-Variance Decomposition**: The expected test MSE can be decomposed as:
   $$\mathbb{E}[(y_0 - \hat{f}(x_0))^2] = \text{Var}(\hat{f}(x_0)) + [\text{Bias}(\hat{f}(x_0))]^2 + \text{Var}(\epsilon)$$
   where $\text{Var}(\epsilon)$ is the irreducible error.
