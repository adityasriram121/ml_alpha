## Mathematical Objectives

1. **Hypothesis Function**:
   $$\hat{y} = X\theta$$
   where:
   * $X$ is the feature matrix of shape $(m, n)$ (including the bias column of $1$s).
   * $\theta$ is the parameter vector of shape $(n, 1)$.

2. **Mean Squared Error (MSE)**:
   $$J(\theta) = \frac{1}{2m} \sum_{i=1}^m (\hat{y}^{(i)} - y^{(i)})^2 = \frac{1}{2m} (X\theta - y)^T (X\theta - y)$$
   *Note: We include the factor of $2$ in the denominator to simplify the derivative when calculating gradients.*

3. **Gradient of MSE with respect to $\theta$**:
   $$\nabla_\theta J(\theta) = \frac{1}{m} X^T (X\theta - y)$$

4. **Normal Equation (Closed-form Solution)**:
   $$\theta = (X^T X)^{-1} X^T y$$
